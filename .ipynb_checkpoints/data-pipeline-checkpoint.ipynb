## 4. Generating the graphs

```python
df.columns
```

```python
#creating the Heatmap for correlation of attributes 

#setting the theme
sns.set_theme(style= 'whitegrid')
plt.figure(figsize = (20,16))
corr_matrix = df.corr(numeric_only = True)
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt= '.2f')
plt.title('1. Correlation Heatmap of Delay Factors')
plt.show()
```

```python
# 2. Histogram : Distribution of arrival delays
plt.figure(figsize=(10, 6))
# Focusing on delays < 200 mins to see the distribution clearly
sns.histplot(df[df['arr_delay'] < 200]['arr_delay'], kde=True, color='teal')
plt.title('2. Distribution of Arrival Delays (minutes)')
plt.show()
```

```python
# 3. Bar plotplot
plt.figure(figsize=(12, 6))

# Sorting to find the 'King of Delays'
order = df.groupby('carrier')['arr_delay'].mean().sort_values(ascending=False).index

sns.barplot(
    data=df, 
    x='carrier', 
    y='arr_delay', 
    hue='carrier',     # Assigning x to hue
    palette='viridis', 
    order=order,
    legend=False       # Hides the redundant legend
)

plt.title('3. Average Arrival Delay per Airline Carrier')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Box Plot: Distribution of Delay Causes
plt.figure(figsize=(12, 6))
delay_causes = ['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']
sns.boxplot(data=df[delay_causes])
plt.title('4. Statistical Spread of Different Delay Causes')
plt.ylabel('Minutes')
plt.show()
```

```python
# 5. Scatter Plot: Late Aircraft Delay vs Total Arrival Delay
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='late_aircraft_delay', y='arr_delay', alpha=0.4, color='darkred')
plt.title('5. Impact of Cascading Delays (Late Aircraft) on Total Delay')
plt.show()
```

```python
#6. Monthly Trends - Line chart
monthly_delays = df.groupby('month')['arr_delay'].mean().reset_index()

plt.figure(figsize=(10, 6))

# Create a line plot to show the trend
sns.lineplot(data=monthly_delays, x='month', y='arr_delay', marker='o', color='royalblue', linewidth=2.5)

# Adding labels and title
plt.title('6. Average Arrival Delay by Month', fontsize=14)
plt.xlabel('Month (1 = January, 12 = December)', fontsize=12)
plt.ylabel('Average Arrival Delay (Minutes)', fontsize=12)

# Ensure all months are shown on the X-axis
plt.xticks(range(1, 13))

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### 5. Insights

#### 1.Which airline is the "king of delays"?
Looking at the Bar Plot (Graph 3), the "King of Delays" is usually the carrier with the highest average bar - WN

Insight: Smaller regional carriers or specific budget airlines often show higher average delays compared to major carriers because they have fewer backup planes when things go wrong.

#### 2. Is weather really the biggest cause, or is it "Late Aircraft"?
Look at the Box Plot (Graph 4) and the Correlation Heatmap (Graph 1)- Late Aircraft

Insight: While major weather events get the most news coverage, "Late Aircraft Delay" is almost always the bigger cause. This is a "cascading effect"—when one flight arrives late, it creates a domino effect that makes all subsequent flights for that plane late. The Heatmap will show a much higher correlation ($> 0.80$) for late_aircraft_delay than for weather_delay.

#### 3. Do delays get worse during specific months?
(To see this, Looking at the figure 6) - Month 7

Insight: Delays typically peak in December (holiday travel and winter weather) and June/July (summer thunderstorms and high passenger volume). If your data shows a spike in month 12, it confirms that seasonal travel demand is a major driver of airline inefficiency.

#### 4: What is the most common experience for a passenger regarding flight delays?

By looking at the Distribution of Arrival Delays (Histogram), we see a huge peak (mode) between 0 and 15 minutes.

Insight: The vast majority of flights arrive either on time or with a very minor delay.
